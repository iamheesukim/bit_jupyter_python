```python
import pandas as pd
from pandas import DataFrame
```

```python
# txt문서를 읽어 DataFrame 구하기
#             읽어올 파일명                   구분기호
dataFrame = pd.read_table('testFile/test_text_file.txt', '|', engine='python')
display(dataFrame)
```
